### We will create the function that translates the results and appropriate form

For that will consider the dataframe that holds the initial results (src_df) and the dataframe that will hold the final results (dest_df). The *dest_df* is given as parameter, modified and returned.

```python
def add_one_by_one(src_df, dest_df, relation=None):
    title = 'Standard pairing'
    if relation != None:
            for index, row in src_df.iterrows():
                if "F1" in row['label']:
                    rn = src_df.iloc[index-1,0]
                    if relation != None and relation in rn:
                        if "False Positives" in src_df.iloc[index-2,0]:
                            arow = ['1By1' + title, src_df.iloc[index-1,0], src_df.iloc[index,1], src_df.iloc[index+1,1], src_df.iloc[index+2,1], src_df.iloc[index+3,1], src_df.iloc[index+4,1], src_df.iloc[index+5,1]]
                            dest_df.loc[len(dest_df)] = arow
                        elif "False Positives" in src_df.iloc[index-3,0]:
                            title = src_df.iloc[index-2,0]
                            if not pd.isnull(src_df.iloc[index-2,1]): 
                                title = title + str(src_df.iloc[index-2,1])
                            arow = ['1By1' + title, src_df.iloc[index-1,0], src_df.iloc[index,1], src_df.iloc[index+1,1], src_df.iloc[index+2,1], src_df.iloc[index+3,1], src_df.iloc[index+4,1], src_df.iloc[index+5,1]]
                            dest_df.loc[len(dest_df)] = arow
                        else:
                            title = src_df.iloc[index-2,0]
                            if not pd.isnull(src_df.iloc[index-2,1]): 
                                title = title + str(src_df.iloc[index-2,1])
                    else:
                        title = src_df.iloc[index-2,0]
                        if not pd.isnull(src_df.iloc[index-2,1]): 
                                title = title + str(src_df.iloc[index-2,1])
                            
    if relation == None:
            for index, row in src_df.iterrows():
                if "F1" in row['label']:
                    if "False Positives" in src_df.iloc[index-2,0]:
                        arow = [title, src_df.iloc[index-1,0], src_df.iloc[index,1], src_df.iloc[index+1,1], src_df.iloc[index+2,1], src_df.iloc[index+3,1], src_df.iloc[index+4,1], src_df.iloc[index+5,1]]
                        dest_df.loc[len(dest_df)] = arow
                    elif "False Positives" in src_df.iloc[index-3,0]:
                        title = src_df.iloc[index-2,0]
                        if not pd.isnull(src_df.iloc[index-2,1]): 
                            title = title + str(src_df.iloc[index-2,1])
                        arow = [title, src_df.iloc[index-1,0], src_df.iloc[index,1], src_df.iloc[index+1,1], src_df.iloc[index+2,1], src_df.iloc[index+3,1], src_df.iloc[index+4,1], src_df.iloc[index+5,1]]
                        dest_df.loc[len(dest_df)] = arow
                    else:
                        title = src_df.iloc[index-2,0]
                        if not pd.isnull(src_df.iloc[index-2,1]): 
                            title = title + str(src_df.iloc[index-2,1])
                        arow = [title, src_df.iloc[index-1,0], src_df.iloc[index,1], src_df.iloc[index+1,1], src_df.iloc[index+2,1], src_df.iloc[index+3,1], src_df.iloc[index+4,1], src_df.iloc[index+5,1]]
                        dest_df.loc[len(dest_df)] = arow
    return dest_df
```

### transform relations, one by one using the function. Put the results in seedev_all

```python
import os.path

new_cols = ['Pairing', 'Eval', 'F1', 'Recall', 'Precision', 'Predictions', 'False Negative', 'False positive']
seedev_all_df = pd.DataFrame(columns=new_cols)

seedev_file = 'SEEDEV-binary/models/model-SEEDEV_ALL/SEEDEV_ALL.csv'
seedev_df = pd.read_csv(seedev_file, sep=':', names = ['label', 'eval'])

seedev_all_df = add_one_by_one(src_df=seedev_df, dest_df=seedev_all_df, relation=None)

seedev_all_df
```

```python
relation_list = 'SEEDEV-binary/seedev_relation_list.csv'
relation_list_df = pd.read_csv(relation_list, sep=':', names = ['relation'])

for index, row in relation_list_df.iterrows():
    a_relation_file =  'SEEDEV-binary/models/model-' + row['relation'] + '/' + row['relation'] + '.csv'
    if os.path.exists(a_relation_file) : 
        a_relation_df = pd.read_csv(a_relation_file, sep=':', names = ['label', 'eval'])
        seedev_all_df = add_one_by_one(src_df=a_relation_df, dest_df=seedev_all_df, relation=row['relation'])
    
    
seedev_all_df.tail(50)
```
